# 1. Busca los dominios del diccionario y los imprime. Lo usamos para sacar todos los dominios que hay en el diccionario de oxford.

```python
def find_domains(f_read, f_write):
  try:
    for line in f_read:
      for part in line.split():
        if "DOM:" in part:
          end = len(line) - 1
          linea = list(line[0:end])
          for idx in range(len(linea)):
            if linea[idx] is 'D' and linea[idx+1] is 'O' and linea[idx+2] is 'M':
              dominio = line[idx+5:end]
              break
          print(dominio)
          f_write.write(dominio + '\n')
  except:
    print('Something failed in the function find_domains reading from the file {}'.format(f_read))
    f_write.write('Something failed in the function find_domains reading from the file {}\n\n'.format(f_read))
    return {}
```

```python
def domains_per_word():
  fileR = open("/content/drive/MyDrive/Colab Notebooks/TFG/Dictionary/#_dictionary.txt", "r") #"r" si queremos leer
  fileW = open("#_domains.txt", "w") #"w" si queremos sobreescribir, "a" si queremos escribir a continuacion del contenido que ya tenga
  find_domains(fileR, fileW)
  fileR.close()
  fileW.close()
```
