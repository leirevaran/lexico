1.1 Letra bat emanda, oxford online hiztegian letra horrekin hasten diren hitzak bilatu eta inprimatzen ditu. hiztegia_lortu_orria() funtziora deitzen dio hau lortzeko.

```python
def hiztegia_lortu_online(letra, fichero):
  try:
    URL = 'https://www.lexico.com/list/' + letra
    html = urlopen(URL).read().decode('utf-8')
  except HTTPError:
    print('Errorea {} letraren weborrian sartzerakoan'.format(letra))
    return {}
  
  try:
    html_info = BeautifulSoup(html, "lxml") #zuhaitz bat 'html'aren informazioarekin
    for div in html_info.find_all(attrs={"class": "breadcrumbs"}):
      div.decompose() #kentzen dugu zuhaitzetik interesatzen ez zaiguna
    
    orri_zenbakiak_aux = html_info.find(attrs={"class": "sitemap_pagination"})
    orri_zenbakiak = orri_zenbakiak_aux.find(attrs={"class": "pagination"})

    if orri_zenbakiak == None: #letrak orrialde bakarra badauka
      result = hiztegia_lortu_orria(letra, '1', fichero)
      result -=1
    else:
      azken_orria = orri_zenbakiak.find(attrs={"class": "last"})
      azken_orria_string = str(azken_orria)
      azken_zenbakia = re.split('<a href="/list/(.)/(.*?)">Last »</a>', azken_orria_string)
      orria = int(azken_zenbakia[2]) #azken orriaren zenbakia
      lista = list(range(1, orria + 1))
    
      #hitz_kopurua = hiztegia_lortu_orria(letra, '1', fichero)

      for idx in lista:
        try:
          hitz_kopurua = hiztegia_lortu_orria(letra, str(idx), fichero)
        except:
          print('ops! could not call the function hiztegia_lortu_orria with the letter {}'.format(letra))
          fichero.write('ops! could not call the function hiztegia_lortu_orria with the letter {}\n\n'.format(letra))    
    
      result = 300*(orria-1) + hitz_kopurua - 1
             
    print('TOTAL AMOUNT OF WORDS OF THE LETTER {}:'.format(letra), end=" ")
    print(result)

  except:
    print('Errorea hiztegia_lortu_orria funtziora deitzerakoan'.format(letra))
    return {}
```

1.2 Letra eta zenbaki bat emanda, oxford online hiztegian letra horrekin hasten diren eta zenbaki horren orrian agertzen diren hitzak bilatu eta inprimatzen ditu.

```python
def hiztegia_lortu_orria(letra, orria, fichero):
  try:
    URL = 'https://www.lexico.com/list/' + letra + '/' + orria #karaktere bereziak: letra=0
    html = urlopen(URL).read().decode('utf-8')
  except HTTPError:
    print('Errorea {} letraren weborrian sartzerakoan'.format(letra))
    return {}
  
  try:
    html_info = BeautifulSoup(html, "lxml") #zuhaitz bat 'html'aren informazioarekin
    for div in html_info.find_all(attrs={"class": "breadcrumbs"}):
      div.decompose() #kentzen dugu zuhaitzetik interesatzen ez zaiguna

    hitzen_zerrenda = html_info.find(attrs={"class": "textBlock"})
    hiztegia = hitzen_zerrenda.find_all('li', "")
    
    i = 1 #hitzaren indizea

    for hz in hiztegia:
      if str(hz)[:4] == '<li>':
        hz_string = str(hz) #hau ez kendu
        hitzak = re.split('<a href="/definition/(.*?)">(.*?)</a>', hz_string)

        if i<10:
          print('LETTER: {} '.format(letra) + 'PAGE: {} '.format(orria) + 'NUMBER: {}   '.format(i) + 'WORD: {}'.format(hitzak[2]))
          fichero.write('LETTER: {} '.format(letra) + 'PAGE: {} '.format(orria) + 'NUMBER: {}   '.format(i) + 'WORD: {}\n'.format(hitzak[2]))  
        elif (i<100):
          print('LETTER: {} '.format(letra) + 'PAGE: {} '.format(orria) + 'NUMBER: {}  '.format(i) + 'WORD: {}'.format(hitzak[2]))
          fichero.write('LETTER: {} '.format(letra) + 'PAGE: {} '.format(orria) + 'NUMBER: {}  '.format(i) + 'WORD: {}\n'.format(hitzak[2]))
        else:
          print('LETTER: {} '.format(letra) + 'PAGE: {} '.format(orria) + 'NUMBER: {} '.format(i) + 'WORD: {}'.format(hitzak[2]))
          fichero.write('LETTER: {} '.format(letra) + 'PAGE: {} '.format(orria) + 'NUMBER: {} '.format(i) + 'WORD: {}\n'.format(hitzak[2]))


        i +=1

    print('\n\n')
            
    return int(i)

  except:
    print('Errorea {} letraren hiztegia lortzerakoan'.format(letra))
    return {}
```
